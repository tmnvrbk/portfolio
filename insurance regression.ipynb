```python
# Importing all the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Loading the insurance document into a dataframe with pandas
df = pd.read_csv("insurance.csv")
df.head()
```

```python
# Assigning age to x and charges to y
x = df.iloc[:,:1].values
y = df.iloc[:,6:7].values

# Plotting a scatter graph with the above values
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Fitting age and charges into a linear regression model

insurance_model = LinearRegression()
insurance_model.fit(x, y)
```

```python
# Plotting the same scattergraph but now adding the linear regression prediction inside of it

y_pred = insurance_model.predict(x)

plt.scatter(x,y)
plt.plot(x,y_pred, color = 'r')
plt.show()
```

```python
# Predict an unknown value for age 70
unk_x = [[70]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

# Printing the same scattergraph with the prediction now added as well
print("The charges for a 70 year old would be:", insurance_model.predict(unk_x))
plt.scatter(x,y)
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

```
